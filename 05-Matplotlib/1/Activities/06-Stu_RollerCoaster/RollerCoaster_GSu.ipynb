```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create the X and Y axis lists
time = np.arange(0,130,10)
speed_chain = [9, 8, 90, 85, 80, 70, 70, 65, 55, 60, 70, 65, 50]
speed_launch = [75, 70, 60, 65, 60, 45, 55, 50, 40, 40, 35, 35, 30] # Dependencies
```

```python
# Use dots or other markers for your plots, and change their colors
plt.plot(time, speed_chain, linewidth=1, color="red", label="Chain")
plt.plot(time, speed_launch, linewidth=1, color="blue", label="launch")
```

```python
# Add labels to the x and y axes
plt.title("Juxtaposed Chain Ride and Launch Ride")
plt.xlabel("Ride Length (Secods)")
plt.ylabel("Speed (MPH)")
```

```python
# Adds a legend and sets its location to the lower right
plt.legend(loc="best")
```

```python
# Set a grid on the plot
plt.grid()
```

```python
# Save the plot and display it
# plt.savefig("sin_cos_with_markers.png")
plt.show()
```
