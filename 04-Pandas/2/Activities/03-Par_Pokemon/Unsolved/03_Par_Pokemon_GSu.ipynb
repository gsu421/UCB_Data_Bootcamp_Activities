```python
# Dependencies
# Import Dependencies
import pandas as pd
import os
```

```python
# File to load
# Create a reference the CSV file desired
csv_path = os.path.join('..', 'Resources', 'Pokemon.csv')
```

```python
# Read with Pandas

# Read the CSV into a Pandas DataFrame
pokemon_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
pokemon_df.head()
```

```python
# Create new dataframe
pokemon_clean_df = pd.DataFrame(pokemon_df,columns=['Type 1', 'HP', 'Attack', 'Defense','Sp. Atk', 'Sp. Def', 'Speed'])
pokemon_clean_df.head()
```

```python
# Create the GroupBy object based on the "Type 1" column
grouped_type1 = pokemon_clean_df.groupby('Type 1')

# Calculate averages for combat stats using the .mean() method
grouped_mean = grouped_type1.mean()
grouped_mean.head()


```

```python
# Total number of points
grouped_mean['Total'] = grouped_mean.sum(axis=1)
grouped_mean.head()
```
